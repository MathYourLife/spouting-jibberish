## Exponential Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### The Maths

#### Cumulative Distribution Function
$$
\text{F}(x) = 1 - e^{-\lambda x}
$$

```python
def cdf(lam, x):
    return 1 - np.exp(-lam * x)

def plot_cdf(lam, y_scale=1):
    x = np.linspace(0,1,100)
    y = [cdf(lam, xi) * y_scale for xi in x]
    plt.plot(x, y, label="$\lambda=%g$" % lam, lw=2)

[plot_cdf(lam) for lam in [8.6, 3.6, 1.6, 0.1]]
plt.xlabel("x")
plt.ylabel("$Quantile\ F(x)$")
plt.legend(loc='lower right')
plt.grid()
```

#### Probability Density Function
$$
\text{pdf}(x) = \frac{d}{dx} F(x) = \lambda e^{-\lambda x}
$$

```python
def pdf(lam, x):
    return lam * np.exp(-lam * x)

def plot_pdf(lam):
    x = np.linspace(0,1,100)
    y = [pdf(lam, xi) for xi in x]
    plt.plot(x, y, label="$\lambda=%g$" % lam, lw=2)

[plot_pdf(lam) for lam in [8.6, 3.6, 1.6, 0.1]]
plt.xlabel("x")
plt.ylabel("$pdf(x)$")
plt.legend()
plt.grid()
```

### Generating an Exponentially Distributed Data Stream

A data stream can be generated from a arbitrary distribution with a combination of a uniform random number generator (RNG) and the cumulative distribution function $F(x)$.  The uniform RNG, can generate a value on the range $[0, 1)$ which can be translated into the desired quantile of the distribution.  For example, if the RNG generates a 0.5 this corresponds to the 50th percentile, the median, of the distribution.  Conversion from the 0.5 to the distributions median value is accomplished with calculating the inverse of the cumulative distribution function.
$$
F(x) = 1 - e^{-\lambda x} \\
Quantile = 1 - e^{-\lambda x} \\
e^{-\lambda x} = 1 - Quantile \\
-\lambda x = \ln{ \left( 1 - Quantile \right) } \\
x = \frac{1}{-\lambda}\ln{ \left( 1 - Quantile \right) } \\
$$
Note: A simplification of $x = \frac{1}{-\lambda}\ln{ \left(Quantile \right) }$ could be used.  This will not be used in this demonstration as it also changes the quantile selected to $100\% - quantile$.

```python
def inv_cdf(lam, quantile):
    return 1./-lam * np.log(1 - quantile)
```

Below shows a quantile of 0.5 being selected and then mapped to the appropriate $x$ value in the distribution.

```python
lam = 3.6

x = np.linspace(0,1,100)
y = [cdf(lam, xi) for xi in x]
plt.plot(x, y, label="$\lambda=%g$" % lam, lw=2)

def map_to_cdf(yi, alpha=1):
    xi = inv_cdf(lam, yi)
    plt.arrow(0, yi, xi, 0, alpha=alpha, color='red', ls='dashed', length_includes_head=True)
    plt.arrow(xi, yi, 0, -yi, alpha=alpha, color='red', ls='dashed', length_includes_head=True)

map_to_cdf(0.5)
plt.xlabel("x")
plt.ylabel("$pdf(x)$")
plt.legend(loc='lower right')
plt.grid()
```

```python
x = np.linspace(0,1,100)
y = [cdf(lam, xi) for xi in x]
plt.plot(x, y, label="$\lambda=%g$" % lam, lw=2)

[map_to_cdf(val, 0.2) for val in np.random.rand(150)]

plt.xlabel("x")
plt.ylabel("$pdf(x)$")
plt.grid()
```

by uniformly selecting points on the y-axis, the associated values on the x-axis are the generated exponentially distributed data set.

```python
def gen_exponential(lam, N):
    for _ in range(N):
        yield inv_cdf(lam, np.random.rand())
```

```python
N=100

x = np.array([val for val in gen_exponential(lam, N)], dtype=np.float)
plt.xlim([0,1])
# To keep the plots consistent, only plot values from 0 to 1
plt.hist(x[x < 1], bins=20, normed=True, alpha=0.4, label='simulated $\lambda$=%g' % lam)
plot_pdf(lam)
plt.legend()
plt.grid()
```

So far, we've explored the mathematics behind the exponential distribution, converted it to a exponential random number generator and verified that the generated distribution appears to match the theoretical distribution.

### Fitting Exponentially Distributed Data

Previously we've worked from a known $\lambda$ term to create a distribution.  Now given a distribution, we'd like to obtain the maximum likelihood estimate $\hat{\lambda}$. [statlect.com](http://www.statlect.com/exponential_distribution_maximum_likelihood.htm) has a nice walk through on derivation of the estimate which becomes the reciprocal of the sample mean.
$$
\hat{\lambda} = \frac{N}{\sum_{i=1}^N x_i} = 1/\bar{X}
$$

```python
lambda_est = 1/np.mean(x)
print("estimate for lambda=%g which is %g%% off the true value of %g" % (lambda_est, (lambda_est-lam)/lam*100, lam))
```

## How Good of A Fit?

Now that we have created the most likely estimate for lambda from the data, the final step is to quantify how well the estimated distribution matches the observed.  One such method is the Kolmogorov-Smirnov (KS) Test which compares the expected cumulative distribution function with the observed empirical cumulative distirbution function.  Recall from above that the cdf for an exponential distribution is:

```python
plot_cdf(lam)
plt.xlabel("x")
plt.ylabel("$Quantile\ F(x)$")
plt.legend(loc='lower right')
plt.grid()
```

To illustrate the comparison, we'll change the y-axis from the quantile into the number of samples.  So a y-axis value of 600 can be interpreted as:

"For 100 independent and identically distributed (iid) random samples, we would expect about 60 of them to have a value less than 0.25."

```python
plot_cdf(lam, len(x))
plt.xlabel("x")
plt.ylabel("Number of Samples < x")
plt.legend(loc='lower right')
plt.grid()
```

In order to create the observed emperical cdf function, we can sort the data and plot a running tally as the values increase.

```python
def plot_ecdf(x, y_scale=1):
    def gen_x(x):
        yield 0
        for xi in sorted(x):
            yield xi
            yield xi
    def gen_y(x):
        yield 0
        yield 0
        for yi in range(1, len(x) + 1):
            yield float(yi) / len(x) * y_scale
            yield float(yi) / len(x) * y_scale
    xs = [pt[0] for pt in zip(gen_x(x), gen_y(x))]
    ys = [pt[1] for pt in zip(gen_x(x), gen_y(x))]
    plt.plot(xs, ys, label="Observed CDF", lw=2)

plot_cdf(lam, len(x))
plot_ecdf(x[x<1], len(x))
plt.xlabel("x")
plt.ylabel("Number of Samples < x")
plt.legend(loc='lower right')
plt.grid()
```

Converting the y-axis back to quantiles by dividing by the number of samples

```python
plot_cdf(lam)
plot_ecdf(x[x<1])
plt.xlabel("x")
plt.ylabel("$Quantile\ F(x)$")
plt.legend(loc='lower right')
plt.grid()
```

### Kolmogorov-Smirnov

The Kolmogorov-Smirnov test essentially searched for the largest vertical distance between the two distributions.  The larger the gap, the less likely the two distributions are drawn from the same population.



```python
def ks_exp_test(x):
    sx = np.array(sorted(x), dtype=np.float)
    quantiles = np.array([(q+1.)/len(x) for q in range(0,len(x))], dtype=np.float)
    fx = np.array([cdf(lam, xi) for xi in sorted(x)])
    
    # If the largest residual is above the cdf
    D_left = max(np.abs(quantiles-fx))
    # If the largest residual is below the cdf
    D_right = max(np.abs(quantiles[:-1]-fx[1:]))
    return max(D_left, D_right)
```

Because of the step-wise nature of the ecdf if the ecdf is above the cdf the largest distance will be found at the start of the step.  If the ecdf is below the cdf, the largest distance will be found at the end of the step.

```python
D = ks_exp_test(x)
print("There is 1 point in the sample that is %g away from its expected quantile" % D)
```

This calculation can also be done with the scipy package which will return an associated p-value for hypothesis testing

```python
from scipy import stats
D, p_value = stats.kstest(x, 'expon', args=(0, 1./lam))
print(D, p_value)
```
